```python
source_file = "input1.txt"
with open(source_file) as f:
    lines = f.readlines()
    
reports = []
for line in lines:
    reports.append(line.strip().split(" "))
```

```python
def isReportSafe(report):
    safe = True
    distance = 0

    for i in range(len(report)):
        if i == len(report) - 1:
            break
        else:
            itemDistance = int(report[i]) - int(report[i+1])
            # Check no change
            if itemDistance == 0:
                safe = False
                break
            # Check directional change
            elif (distance < 0 and itemDistance > 0) or (distance > 0 and itemDistance < 0):
                safe = False
                break
            # Check too much change
            elif itemDistance > 3 or itemDistance < -3:
                safe = False
                break
            else:
                distance += itemDistance
    
    print(f"{report} is {'safe' if safe else 'unsafe'}")
    return safe

safeCount = 0
for report in reports:
    if isReportSafe(report):
        safeCount += 1
print(safeCount)
```
